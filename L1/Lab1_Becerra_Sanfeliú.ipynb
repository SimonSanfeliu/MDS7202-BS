Responda las siguientes preguntas (0.2 pts c/u):

1. ¿Qué es `Git` y cómo se relaciona con el control de versiones?

2. Nombre y explique 3 diferencias entre `Git` y `Github`.

3. ¿Qué es un `hash` y cómo se representa en un `commit`? De un ejemplo.

4. ¿Que elementos componen a la estructura de datos de `Git`?

5. ¿Que comandos de `Git` sirven para crear un repositorio y hacer seguimiento a cambios en los archivos?

1. `Git` es un sistema de control de versiones distribuido, lo que implica que se van controlando las distintas versiones de distintos códigos de un proyecto en distintos dispositivos al mismo tiempo (en computadores locales y en un servidor).

2. 
   * `Git` es un software, `Github` es un servicio: `Git` es un programa instalado en un ordenador en el cual se ejecutan comando, mientras que Github resulta ser un servicio que interactua y almacena repositorios de `Git`.
   * `Git` es una herramienta de comando, `Github` es una herramienta de interfaz gráfica: `Git`, al ser un software, interactua con los repositorios a través de comando en la terminal, mientras que `Github` es una interfaz gráfica que permite interactuar con los repositorios de Git de una manera más amigable y atractiva.
   * `Git` crea repositorios locales para trackear cambios locales, `Github` es un espacio en la nube para almacenar repositorios: `Git` es capaz de generar repositorios locales en el ordenador en el cual se realizan los cambios, trackeando estos cambios en un mismo ordenador, mientras que `Github` es un servicio que permite almacenar repositorios en la nube al generar una copia de estos.

3. En el contexto de `Git` y `Github`, `hash` es un código alfanumérico de 40 caracteres que permite identificar distintos conjuntos de datos. `Git` utiliza el hash SHA1. Este código puede apuntar a un `blob`, `tree` o `commit`, pudiendo identificar distintas versiones de estos con sus códigos `hash` respectivos o con tan sólo con los primeros términos de dichos códigos. Un ejemplo de esto sería el código `c8b1a62d7299552b0654f930d695b33109214111` de más adelante en el laboratorio.

4. La estructura de datos de `Git` se compone priuncipalmente de 4 objetos: `blobs`, `trees`, `commits` y `hashes`. Los `blobs` son los contenidos de los archivos almacenados y se genera un `blob` por cada modificación del cada archivo que se almacene en el repositorio. Los `trees` son directorios para los `blobs` y otros `trees` (análogo a las carpetas de un ordenador), habiendo siempre un `tree` en la raíz del repositorio. Los `commits` son los registros de los cambios realizados en el repositorio, apuntan a un `tree` en específico, mostrando cómo lucía el proyecto en un cierto punto en el tiempo. Por último, los `hashes` son identificadores únicos de cada objeto de `Git`, los cuales se generan a partir de los contenidos de los objetos, por ende, sirven para dirigirse a un `blob`, `tree` o `commit` en específico.

5. Para crear un repositorio, se debe usar el comando `git init /ruta/del/directorio`. Por su parte, para hacer el seguimiento de los cambios en los archivos se puede usar `git log` para ver el historial de `commits` hechos al repositorio o también se puede usar `git status` para ver el estado de todos los archivos del repositorio, tanto aquellos que estén en un `commit` como aquellos que aún no son agregados a uno.

Responda si las siguientes afirmaciones son Verdaderas (V) o Falsas (F). Justifique en **cualquiera** de los casos (0.2 pts c/u):


6. `Git` siempre resuelve automáticamente los problemas de lineas al hacer merge entre diferentes ramas.
7. Ejecutar `git pull` sobre un repositorio es equivalente a ejecutar `git clone`.
8. `git status` genera los mismos resultados que `git log`.
9. En `Git` es posible deshacer cambios y volver a versiones anteriores del código.
10. El comando `git commit` se utiliza para enviar cambios al repositorio remoto.

6. Falso. Los problemas de líneas de códigos entre distintas ramas deben resolverse manualmente, pero `Git` detectará los conflictos y los mostrará al ocurrirse.

7. Falso. El comando `git pull` trae las versiones de los códigos del proyecto que están en el repositorio remoto al repositorio local, mezclándose éstas con las que se tengan localmente. Para que este comando funcione, ya se debe tener el repositorio local que se conecta con el remoto. En cambio, `git clone` clona el repositorio, siendo este comando usado para traer por primera vez el repositorio al ambiente local.

8. Falso. El comando `git status` entrega el estado de todos los archivos del repositorio, tanto aquellos que estén declarados en `commits` como los que no. Por su parte, `git log` sólo entrega el historial de `commits`.

9. Verdadero. Al ser un sistema de control de versiones, es posible volver a versiones anteriores que sean más estables que las actuales y poder deshacer cambios que resulten conflictivos o ineficientes.

10. Falso. Para enviar cambios al repositorio remoto se utiliza `git push`, pero lo que se manda con dicho comando son los `commits` hechos con `git commit`.

Por último, investigue y comente para qué sirven los siguientes comandos:

11. `git stash` (0.25 puntos)
12. `git reset`. ¿Cuál es la diferencia de este comando con `git revert`?. (0.25 puntos)

11. El comando `git stash` almacena temporalmente cualquier cambio en el código en un *stash* (escondite). Coge los cambios que no han sido confirmado, los guarda aparte y los limpia del directorio de trabajo, permitiendo que el usuario tenga ahora la libertad de cambiar de rama, hacer cambios o cualquier acción que quiere ejecutar sobre el repositorio. Cuando el usuario quiera volver a los cambios que guardó, puede usar `git stash apply` para recuperarlos O `git stash pop` para recuperarlos y eliminarlos del stash. Como se aprecia, este comando sirve principalmente por si se necesita cambiar rápidamente de contexto y no quieres perder los cambios que has hecho en el código pero tampoco quieres hacer un commit al proyecto todavía.

12. El comando `git reset <archivo>` quita los cambios del área de *staging*, en donde se definen todos los cambios que irán en el `commit`. Esto lo hace dejando el repositorio trabajado sin cambios, por lo que el archivo al ser *unstaged* no sobreescribe ningún cambio. Por su parte, `git revert <commit>` deshace todos los cambios que fueron *staged* en el `commit` especificado y luego estos se aplican a la rama en la que se esté trabajando.
